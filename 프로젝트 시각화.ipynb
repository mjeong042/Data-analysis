```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
customer_df = pd.read_csv('data/customer_data.csv', sep='\t')
customer_df
```

```python
customer_df.isna().sum()
```

```python
customer_df[customer_df.isna().any(axis=1)]
```

```python
customer_df['age_group'] = (2025 - customer_df['birth_year']) // 10 * 10
customer_df['annual_income'] = customer_df.groupby('age_group')['annual_income'].transform(lambda x: x.fillna(x.median()))
```

```python
customer_df.isna().sum()
```

```python
customer_df.describe()
```

```python
customer_df['age_group'].value_counts().sort_index()
```

```python
customer_df = customer_df[customer_df['age_group'] < 100]
```

```python
customer_df['marital_status'].unique()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
customer_df.loc[:, 'marital_status_num'] = le.fit_transform(customer_df['marital_status'])
```

```python
print(dict(zip(le.classes_, le.transform(le.classes_))))
```

```python
customer_df
```

```python
customer_df[customer_df.duplicated()]    #중복값없음 확인함.
```

```python
promotion_cols = [f'promotion_{i}' for i in range(1, 7)]
customer_df['promotion_total'] = customer_df[promotion_cols].sum(axis=1)
```

```python
spending_cols = ['amount_alcohol', 'amount_fruit', 'amount_meat', 'amount_fish', 'amount_snack']
customer_df['total_spent'] = customer_df[spending_cols].sum(axis=1)
```

```python
customer_df[['promotion_total', 'total_spent']].head()
```

```python
top_customers = customer_df.sort_values('total_spent', ascending=False).head(20)
top_customers
```

```python
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.boxplot(x='age_group', y='total_spent', data=customer_df)
plt.xlabel('Age Group')
plt.ylabel('Total Spent')
plt.title('Total Spent by Age Group')
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = customer_df.corr(numeric_only=True)
correlation_with_spent = correlation_matrix['total_spent'].sort_values(ascending=False)



plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
grouped = customer_df.groupby('age_group')[[
    'amount_alcohol', 'amount_fruit', 'amount_meat', 'amount_fish', 'amount_snack'
]].mean()
```

```python
grouped.plot(kind='bar', figsize=(12, 6))
plt.title('나이대별 주요 구매 품목 평균 지출액')
plt.xlabel('나이대 (age_group)')
plt.ylabel('평균 지출액')
plt.xticks(rotation=0)
plt.legend(title='품목')
plt.tight_layout()
plt.show()
```
